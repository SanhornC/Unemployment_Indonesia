```python
# Define the mapping for tk31a based on the provided categories
tk31a_mapping = {
    1: "Agriculture, forestry, fishing and hunting",
    2: "Mining and quarrying",
    3: "Manufacturing",
    4: "Electricity, gas, water",
    5: "Construction",
    6: "Wholesale, retail, restaurants and hotels",
    7: "Transportation, storage and communications",
    8: "Finance, insurance, real estate and business services",
    9: "Social services",
    10: "Activities that cannot be classified"
}

# Map the tk31a column to categorical labels
df["tk31a_category"] = df["tk31a"].map(tk31a_mapping)

# Display the filtered dataset with the new category column
df
```

```python
# Initialize an empty dictionary to store category-wise DataFrames
category_dfs = {}

# Iterate through each category in the mapping
for category in tk31a_mapping.values():
    # Filter the DataFrame for the current category
    category_df = df[df["tk31a_category"] == category]
    
    # Store the filtered DataFrame in the dictionary
    category_dfs[category] = category_df

# Display each category separately
category_df
```

```python
# Save each category as a separate CSV file
output_files = {}
output_dir = "/Users/mingjiaodiao/Desktop/2025 spring/econ 399 lab/Book 3A datasets/different_industry" 

for category, df in category_dfs.items():
    # Generate a filename based on the category name
    filename = f"{category.replace(',', '').replace(' ', '_').replace('/', '_')}.csv"
    file_path = output_dir + filename
    
    # Save the DataFrame to CSV
    df.to_csv(file_path, index=False)
    
    # Store the file path for reference
    output_files[category] = file_path

# Provide the download links for the saved files
output_files
```
