```python
mg2 = pd.read_csv("/Users/sanhorn/Desktop/Internships/Econ Data Lab/IFLS5 Dataset/b3a_mg2.csv")
mg2
```

```python
common_hhids = set(mg1["hhid14_9"].unique()).intersection(set(mg2["hhid14_9"].unique()))
common_mg1 = mg1[mg1["hhid14_9"].isin(common_hhids)]
common_mg2 = mg2[mg2["hhid14_9"].isin(common_hhids)]

common_mg = pd.merge(common_mg1, common_mg2, on=["hhid14_9", "pid14"], how="outer")
common_mg = common_mg.sort_values(by=['hhid14_9', 'pid14'])

not_mg1 = mg1[~mg1["hhid14_9"].isin(common_hhids)]
not_mg2 = mg2[~mg2["hhid14_9"].isin(common_hhids)]

not_mg = pd.concat([not_mg1, not_mg2], sort=False)
not_mg = not_mg.sort_values(by=['hhid14_9', 'pid14'])

all_mg = pd.concat([common_mg, not_mg], ignore_index=True)
all_mg = all_mg.sort_values(by=['hhid14_9', 'pid14'])
all_mg = all_mg.reset_index(drop=True)

all_mg.to_csv('b3_migration.csv', index=False)
```
