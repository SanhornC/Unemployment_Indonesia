```python
selected_columns = ["hhid14_9", "pid14", "tk28year", "tk28"]
df = df[selected_columns]
```

```python
new_df = pd.DataFrame(columns = ["Person", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014"])

df["person"] = df["hhid14_9"].astype(str) + df["pid14"].astype(str)
df = df.drop(columns = ["hhid14_9", "pid14"])

unique_person = df["person"].unique()
new_df["Person"] = unique_person
```

```python
# Pivot the data
df_wide = df.pivot(index='person', columns='tk28year', values='tk28').reset_index()
df_wide.rename(columns={'person': 'Person'}, inplace=True)
df_wide.columns = df_wide.columns.map(str)

# Merge new_df with df_wide
merged_df = pd.merge(new_df, df_wide, on='Person', how='left')

# Remove duplicate columns
for year in range(2007, 2015):
    y = str(year)
    merged_df[y] = merged_df[f"{y}_x"].fillna(merged_df[f"{y}_y"])
    merged_df.drop(columns=[f"{y}_x", f"{y}_y"], inplace=True, errors='ignore')

merged_df.replace({1: 'has job', 3: 'no job'}, inplace=True)

merged_df.to_csv('/Users/sanhorn/Desktop/Internships/Econ Data Lab/IFLS5 Dataset/tk28_cleaned.csv', index=False)

    
```

```python
path = "/Users/sanhorn/Desktop/Internships/Econ Data Lab/IFLS5 Dataset/IFLS4/b3a_tk3.csv"
selected_columns = ["hhid07_9", "pid07", "tk28year", "tk28"]

def employment_visualize(path, selected_columns):
    df = pd.read_csv(path)
    df = df[selected_columns]
    new_df = pd.DataFrame(columns = ["Person", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007"])
    df["person"] = df["hhid07_9"].astype(str) + df["pid07"].astype(str)
    df = df.drop(columns = ["hhid07_9", "pid07"])
    unique_person = df["person"].unique()
    new_df["Person"] = unique_person
    df_wide = df.pivot(index='person', columns='tk28year', values='tk28').reset_index()
    df_wide.rename(columns={'person': 'Person'}, inplace=True)
    df_wide.columns = df_wide.columns.map(str)
    
    merged_df = pd.merge(new_df, df_wide, on='Person', how='left')
    for year in range(1999, 2008):
        y = str(year)
        merged_df[y] = merged_df[f"{y}_x"].fillna(merged_df[f"{y}_y"])
        merged_df.drop(columns=[f"{y}_x", f"{y}_y"], inplace=True, errors='ignore')
    merged_df.replace({1: 'has job', 3: 'no job'}, inplace=True)
    return merged_df


new_df = employment_visualize(path, selected_columns)
new_df.to_csv('/Users/sanhorn/Desktop/Internships/Econ Data Lab/IFLS5 Dataset/04_tk28_cleaned.csv', index=False)
```
