```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/sanhorn/Desktop/Internships/Econ Data Lab/province_unemploymentrate/employed_by_province.csv')
df_copy = df.copy()
```

```python
provinces = df['Province'].tolist()
provinces
```

```python
province_to_geo_unit = {
    # Sumatra
    'Aceh': 'Sumatra',
    'North Sumatra': 'Sumatra',
    'West Sumatra': 'Sumatra',
    'Riau': 'Sumatra',
    'Jambi': 'Sumatra',
    'South Sumatra': 'Sumatra',
    'Bengkulu': 'Sumatra',
    'Lampung': 'Sumatra',
    
    # Java
    'DKI Jakarta': 'Java',
    'West Java': 'Java',
    'Central Java': 'Java',
    'DI Yogyakarta': 'Java',
    'East Java': 'Java',
    
    # Nusa Tenggara (Lesser Sunda Islands)
    'Bali': 'Nusa Tenggara (Lesser Sunda Islands)',
    'West Nusa Tenggara': 'Nusa Tenggara (Lesser Sunda Islands)',
    'East Nusa Tenggara': 'Nusa Tenggara (Lesser Sunda Islands)',
    'East Timor': 'Nusa Tenggara (Lesser Sunda Islands)',
    
    # Kalimantan
    'West Kalimantan': 'Kalimantan',
    'Central Kalimantan': 'Kalimantan',
    'South Kalimantan': 'Kalimantan',
    'East Kalimantan': 'Kalimantan',
    
    # Sulawesi
    'North Sulawesi': 'Sulawesi',
    'Central Sulawesi': 'Sulawesi',
    'South Sulawesi': 'Sulawesi',
    'Southeast Sulawesi': 'Sulawesi',
    
    # Maluku Islands
    'Maluku': 'Maluku Islands',
    
    # Papua (Western New Guinea)
    'Irian Jaya': 'Papua (Western New Guinea)',
    
    # Keep Unknown as is
    'Unknown': 'Unknown'
}
```

```python
df_copy['Geographical Unit'] = df_copy['Province'].map(province_to_geo_unit)

# Check for unmapped provinces
unmapped_provinces = df_copy[df_copy['Geographical Unit'].isna()]['Province'].unique()
if len(unmapped_provinces) > 0:
    df_copy['Geographical Unit'] = df_copy['Geographical Unit'].fillna('Other')


years = [str(year) for year in range(2007, 2015)]  
employed_df = df_copy.groupby('Geographical Unit')[years].sum().reset_index()
employed_df

```

```python
df1 = pd.read_csv("/Users/sanhorn/Desktop/Internships/Econ Data Lab/province_unemploymentrate/unemployed_by_province.csv")
df1_copy = df1.copy()

df1_copy['Geographical Unit'] = df1_copy['Province'].map(province_to_geo_unit)

unmapped_provinces = df1_copy[df1_copy['Geographical Unit'].isna()]['Province'].unique()
if len(unmapped_provinces) > 0:
    df1_copy['Geographical Unit'] = df1_copy['Geographical Unit'].fillna('Other')

years = [str(year) for year in range(2007, 2015)]  
unemployed_df = df1_copy.groupby('Geographical Unit')[years].sum().reset_index()
unemployed_df
```

```python
employed_rate_df = employed_df.copy()
unemployed_rate_df = unemployed_df.copy()
employed_rate_df.set_index('Geographical Unit', inplace=True)
unemployed_rate_df.set_index('Geographical Unit', inplace=True)

total = employed_rate_df + unemployed_rate_df
employed_rate_df = employed_rate_df / total
employed_rate_df = employed_rate_df.round(2)
unemployed_rate_df = unemployed_rate_df / total
unemployed_rate_df = unemployed_rate_df.round(2)

employed_rate_df
```
