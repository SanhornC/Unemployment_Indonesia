```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("/Users/sanhorn/Desktop/Internships/Econ Data Lab/province_unemploymentrate/ifls5_migration_panel_by_year.csv")
df
```

```python
# First, let's split the hhid_pid column into hhid and pid
def split_hhid_pid(id_val):
    # Handle potential NaN or non-string values
    if pd.isna(id_val) or not isinstance(id_val, str):
        return None, None
    
    # Split on dash
    parts = id_val.split('-')
    if len(parts) == 2:
        return parts[0], parts[1]
    return id_val, None

# Create new columns for hhid and pid
df['hhid'] = ''
df['pid'] = ''

# Apply the splitting function
for idx, row in df.iterrows():
    hhid, pid = split_hhid_pid(row['hhid_pid'])
    df.at[idx, 'hhid'] = hhid
    df.at[idx, 'pid'] = pid

# Convert hhid and pid to integers if possible
# First, replace any empty strings with NaN
df['hhid'] = df['hhid'].replace('', np.nan)
df['pid'] = df['pid'].replace('', np.nan)

# Now convert to integer and then to string
# This will remove any leading zeros and decimal points

df['hhid'] = df['hhid'].astype(float).astype(int).astype(str)
df['pid'] = df['pid'].astype(float).astype(int).astype(str)


# Create a new clean_id column by concatenating hhid and pid
df['clean_id'] = df['hhid'] + df['pid']

df = df.drop(columns=['hhid_pid', 'hhid', 'pid'])

# Rename the clean_id column to hhid_pid
df = df.rename(columns={'clean_id': 'hhid_pid'})

df

```

```python
# 2. Function to extract the third numeric value from patterns like "62 - 1 - 12 - 62"
def extract_third_value(value):
    if pd.isna(value) or value == 'NA':
        return 'NA'
    
    # Use regex to extract numbers from the string
    numbers = re.findall(r'\d+', str(value))
    
    # If we have at least 3 numbers, return the third one
    if len(numbers) >= 3:
        return numbers[2]
    else:
        return value

# Apply the transformation to all columns except the first one
for col in df.columns[1:]:
    df[col] = df[col].apply(extract_third_value)

# Save the transformed data


print("Transformation complete. Data saved to 'transformed_data.csv'")
df
```

```python
df.to_csv('transformed_province_data.csv', index=False)
```
