```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/sanhorn/Desktop/Internships/Econ Data Lab/IFLS5 Dataset/b3a_tk3.csv")
df
```

```python
tmp_df = df.copy()
selected_columns = ["hhid14_9", "pid14", "tk28year", "tk28", "tk31a"]
tmp_df = tmp_df[selected_columns]

tmp_df["person"] = tmp_df["hhid14_9"].astype(str) + tmp_df["pid14"].astype(str)
tmp_df = tmp_df.drop(columns = ["hhid14_9", "pid14"])
tmp_df

# tk28 indicates has job / no job
# tk31a indicates industry sector
```

```python
province = pd.read_csv("/Users/sanhorn/Desktop/Internships/Econ Data Lab/province_unemploymentrate/transformed_province_data.csv")
province
```

```python
tmp_df['person'] = tmp_df['person'].astype(str)

# Filter tmp_df to only include years between 2007-2014 if needed
# tmp_df = tmp_df[tmp_df['tk28year'].between(2007, 2014)]

# Convert the province dataframe from wide to long format
# This transforms the year columns (2007, 2008, etc.) into a single 'year' column
province_long = pd.melt(
    province,
    id_vars=['hhid_pid'],
    value_vars=[str(year) for year in range(2007, 2015)],  # Or use actual column names if different
    var_name='year',
    value_name='province'
)

# Convert year to integer for proper joining
province_long['year'] = province_long['year'].astype(int)

# Remove rows where province is NaN (years where no province data exists)
# province_long = province_long.dropna(subset=['province'])

# Ensure hhid_pid is in string format for joining
province_long['hhid_pid'] = province_long['hhid_pid'].astype(str)
province_long
```

```python
unique_person_count = tmp_df['person'].nunique()
print(f"Number of unique person IDs in tmp_df: {unique_person_count}")

# Count unique hhid_pid in province_long before merging
unique_hhid_pid_count = province_long['hhid_pid'].nunique()
print(f"Number of unique hhid_pid in province_long: {unique_hhid_pid_count}")

# Count how many unique person IDs from tmp_df exist in province_long
common_ids = set(tmp_df['person'].unique()).intersection(set(province_long['hhid_pid'].unique()))
print(f"Number of common unique IDs between datasets: {len(common_ids)}")
```

```python
# Merge the dataframes
# Left join keeps all rows from tmp_df and only matching rows from province_long
result_df = tmp_df.merge(
    province_long,
    left_on=['person', 'tk28year'],
    right_on=['hhid_pid', 'year'],
    how='inner'  # Changed from 'left' to 'inner'
)

# Drop the redundant columns that came from the merge
result_df = result_df.drop(columns=['hhid_pid', 'year'])

# Display information about the merged dataframe
print(f"Original tmp_df shape: {tmp_df.shape}")
print(f"Result dataframe shape: {result_df.shape}")
print(f"Number of rows with province data: {result_df['province'].notna().sum()}")

# Show some example rows to verify the merge worked correctly
print("\nSample rows from the merged dataframe:")
unique_id = result_df['person'].nunique()
print(unique_id)
result_df

# Save the result if needed
# result_df.to_csv('combined_data.csv', index=False)
```

```python
location_map = {
    11: "Aceh",
    12: "North Sumatra",
    13: "West Sumatra",
    14: "Riau",
    15: "Jambi",
    16: "South Sumatra",
    17: "Bengkulu",
    18: "Lampung",
    31: "DKI Jakarta",
    32: "West Java",
    33: "Central Java",
    34: "DI Yogyakarta",
    35: "East Java",
    51: "Bali",
    52: "West Nusa Tenggara",
    53: "East Nusa Tenggara",
    54: "East Timor",
    61: "West Kalimantan",
    62: "Central Kalimantan",
    63: "South Kalimantan",
    64: "East Kalimantan",
    71: "North Sulawesi",
    72: "Central Sulawesi",
    73: "South Sulawesi",
    74: "Southeast Sulawesi",
    81: "Maluku",
    82: "Irian Jaya",
    99: "Unknown",
}
```

```python
df1 = result_df.copy()
years = ['2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']
employed_by_province = {province_code: {year: 0 for year in years} for province_code in location_map.keys()}
unemployed_by_province = {province_code: {year: 0 for year in years} for province_code in location_map.keys()}

# Create lists to track rows that aren't stored
missed_rows = []
missed_reasons = []

# Process each row in the dataframe
total_rows = 0
stored_rows = 0

for _, row in df1.iterrows():
    total_rows += 1
    year = str(row["tk28year"])
    person_hasJob = row["tk28"]
    person_id = row.get("person", "unknown")  # Get person ID for tracking
    
    # Get province code, use 99 for NaN
    if pd.notna(row["province"]):
        province_code = int(row["province"])
    else:
        province_code = 99
    
    # Check if this row will be stored or skipped
    if province_code not in employed_by_province or year not in years:
        continue
    
    # Count employed and unemployed by province
    stored_rows += 1
    if person_hasJob == 1:  # Person has a job
        employed_by_province[province_code][year] += 1
    else:  # Person doesn't have a job
        unemployed_by_province[province_code][year] += 1

# Create DataFrames
employed_df = pd.DataFrame(employed_by_province).T
unemployed_df = pd.DataFrame(unemployed_by_province).T

# Add province names and set as index
employed_df['Province'] = employed_df.index.map(lambda x: location_map.get(x, f"Unknown {x}"))
employed_df = employed_df.set_index('Province')
unemployed_df['Province'] = unemployed_df.index.map(lambda x: location_map.get(x, f"Unknown {x}"))
unemployed_df = unemployed_df.set_index('Province')

```

```python
# Calculate the total counts (employed + unemployed) for each province and year
total_counts = employed_df + unemployed_df

# Calculate the employed ratio
employed_ratio_df = employed_df / total_counts

# Display the resulting employed ratio table
employed_ratio_df = employed_ratio_df.round(2)
employed_ratio_df
```
