```python
# List of geo unit dataframes and their corresponding names
geo_units = {
    'Java': java,
    'Kalimantan': kalimantan,
    'Maluku': maluku,
    'Nusa Tenggara': nusa_tenggara,
    'Papua': papua,
    'Sulawesi': sulawesi,
    'Sumatra': sumatra
}

# Initialize an empty list to store the reshaped dataframes
reshaped_dfs = []

# Loop through each geo unit dataframe
for geo_unit, df in geo_units.items():
    # Melt the dataframe to reshape it
    melted_df = df.melt(id_vars=['index'], var_name='Year', value_name='Value')
    # Add GeoUnit column
    melted_df['GeoUnit'] = geo_unit
    # Rename columns for clarity
    melted_df.rename(columns={'index': 'Sector'}, inplace=True)
    # Append to the list
    reshaped_dfs.append(melted_df)

# Concatenate all reshaped dataframes into a single dataframe
combined_df = pd.concat(reshaped_dfs, ignore_index=True)

# Pivot the dataframe to have years as columns
pivoted_df = combined_df.pivot_table(
    index=['GeoUnit', 'Sector'],
    columns='Year',
    values='Value'
).reset_index()

# Rename the columns for clarity
pivoted_df.columns.name = None  # Remove the name of the columns
pivoted_df.rename(columns=lambda x: f"Year_{x}" if isinstance(x, str) and x.isdigit() else x, inplace=True)

# Display the pivoted dataframe
pivoted_df
```

```python
pivoted_df.rename(columns=lambda x: x.replace("Year_", "") if x.startswith("Year_") else x, inplace=True)
```
